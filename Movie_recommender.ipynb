```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.rename(columns={"movie_id": "id"}, inplace=True)
```

```python
movies1=movies1[['id','original_title','overview','genres','keywords','cast','crew']]
```

```python
movies1.head()
```

```python
movies1.isnull().sum()
```

```python
movies1.dropna(inplace=True)

```

```python
movies1.duplicated().sum()
```

```python
movies1.shape
```

```python
import ast
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies1['genres'] = movies1['genres'].apply(convert)
movies1.head()
```

```python
display(movies1['genres'].head())
```

```python
movies1['keywords'] = movies1['keywords'].apply(convert)
movies1.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies1['cast'] = movies1['cast'].apply(convert)
movies1.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies1['crew'] = movies1['crew'].apply(fetch_director)
```

```python
movies1.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies1['cast'] = movies1['cast'].apply(collapse)
movies1['crew'] = movies1['crew'].apply(collapse)
movies1['genres'] = movies1['genres'].apply(collapse)
movies1['keywords'] = movies1['keywords'].apply(collapse)
```

```python
movies1.head()
```

```python
movies1['overview'] = movies1['overview'].apply(lambda x:x.split())
```

```python
movies1['tags'] = movies1['overview'] + movies1['genres'] + movies1['keywords'] + movies1['cast'] + movies1['crew']
```

```python
new = movies1.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(movie):
    index = new[new['original_title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].original_title)
```

```python
recommend('Gandhi')
```
